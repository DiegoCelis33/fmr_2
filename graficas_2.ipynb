```python
import numpy as np
import matplotlib.pyplot as plt

data = []

for i in range(1,10):

    data.append(np.loadtxt('exchange6.{a}0/output'.format(a=i), skiprows = 8))
    
data = np.array(data)
```

```python
plt.figure(figsize = (4,4))
    
for i in range(8):
 
    plt.plot(data[1,:,0],data[i,:,3])
```

```python
import scipy.fftpack

psd = []

for i in range(9):
    psd.append(np.abs(scipy.fftpack.fft(data[i,:,1])))
    
    
psd = np.array(psd)
    
f_axis = scipy.fftpack.fftfreq(2000, d=2e-9/2000)
plt.figure(figsize = (4,4))
for ii in range(9):
    plt.plot(f_axis/1e9, psd[ii,:])
    plt.xlim([0, 30])
    plt.ylim([0, 35])
    plt.xlabel('f (GHz)')
    plt.ylabel('Psa (a.u.)')
```

```python
plt.figure(figsize = (4,4))
plt.plot(f_axis/1e9, psd[1,:], marker = 'o')
plt.xlim([0, 30])
plt.ylim([0, 35])
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
plt.grid()
```

```python
plt.figure(figsize = (4,4))
nf_axis = np.delete(f_axis, 20)
npsd = psd[5,:]
npsd= np.delete(npsd, 20)

plt.plot(nf_axis/1e9, npsd, marker = 'o')
plt.xlim([0, 30])
plt.ylim([0, 35])
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
#plt.grid()
```

```python
plt.figure(figsize = (4,4))
d_psd = np.diff(npsd)/np.diff(nf_axis)


plt.figure(figsize = (5,5))
plt.plot(nf_axis[1:]/1e9, d_psd)
plt.xlim([0, 20])
plt.ylim([-1e-8, 1e-8])

plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
```
