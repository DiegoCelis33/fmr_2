```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range(1,11):

    globals()["data" + str(i)] = np.loadtxt("matriz{b}/output".format(b=i), skiprows=9 )
```

```python
lista = []

for i in range(1,11):
    lista.append(np.loadtxt("matriz{b}/output".format(b=i), skiprows=9))
```

```python
lista = np.array(lista)
```

```python
for i in range(10):
    plt.figure(figsize = (9,30))
    plt.subplot(10,3,1+(i*3))
    plt.plot(lista[i,100:,0],lista[i,100:,2])
    plt.subplot(10,3,2+(i*3))
    plt.plot(lista[i,100:,0],lista[i,100:,3])
    plt.subplot(10,3,3+(i*3))
    plt.plot(lista[i,100:,0],lista[i,100:,4])
plt.show

```

```python
mx_avg = np.mean(lista[:,:,2], axis=0)
my_avg = np.mean(lista[:,:,3], axis=0)
mz_avg = np.mean(lista[:,:,4], axis=0)
```

```python
plt.plot(lista[0,:,0], mx_avg)
plt.plot(lista[0,:,0], my_avg)
plt.plot(lista[0,:,0], mz_avg)
```

```python
import scipy.fftpack

psd = (np.abs(scipy.fftpack.fft(mz_avg)))
f_axis = scipy.fftpack.fftfreq(49999, 50e-11/49999)

plt.plot(f_axis/1e9, psd, marker ='o')
plt.xlim([0, 200])
plt.ylim([0,2000])
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
plt.grid()
```

```python
psd = []
for ii in range(10):
    psd.append(np.abs(scipy.fftpack.fft(lista[ii,:,2])))
f_axis = scipy.fftpack.fftfreq(49999, 50e-11/49999)
```

```python
psd = np.array(psd)
```

```python
psd_avg = np.mean(psd, axis = 0)
```

```python
plt.plot(f_axis/1e9, psd_avg, marker ='o')
plt.xlim([0, 200])
plt.ylim([0,2000])
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
plt.grid()
```
