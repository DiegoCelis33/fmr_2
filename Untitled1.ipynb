```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('output')

data2 = np.loadtxt('../time-series_non_magnetic_7/output')
```

```python
plt.figure()
plt.plot(data[:,0],data[:,2])
plt.plot(data2[:,0],data2[:,2])
```

```python
import scipy.fftpack



psd = (np.abs(scipy.fftpack.fft(data[:,1])))
f_axis = scipy.fftpack.fftfreq(2000, d=2e-9/2000)

psd2 = (np.abs(scipy.fftpack.fft(data2[:,1])))
f_axis = scipy.fftpack.fftfreq(2000, d=2e-9/2000)
plt.figure(figsize =(4,4))
plt.plot(f_axis/1e9, psd, label = '2 nm')
plt.plot(f_axis/1e9, psd2, label = '1 nm')
plt.xlim([0, 37.5])
plt.ylim([0, 12])
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
plt.legend()
```

```python
d_psd = np.diff(psd)/np.diff(f_axis)



plt.plot(f_axis[1:]/1e9, d_psd)
plt.xlim([0, 40])
plt.ylim([-2e-9, 2e-9])
plt.grid()
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
```
